```python
import numpy as np

# Step 1: Input data
n = int(input("Enter number of years (including year 0): "))
cash_flows = []

for i in range(n):
    cf = float(input(f"Enter cash flow for year {i}: "))
    cash_flows.append(cf)

rate = float(input("Enter discount rate (e.g. 0.1): "))


# Step 2: NPV Function
def calculate_npv(cash_flows, rate):
    npv = 0
    
    for t in range(len(cash_flows)):
        npv += cash_flows[t] / (1 + rate) ** t
    
    return npv


# Step 3: IRR Function (manual fallback if numpy version fails)
def calculate_irr(cash_flows):
    try:
        return np.irr(cash_flows)
    except:
        # Simple approximation method (trial)
        for r in np.arange(0.0, 1.0, 0.0001):
            npv = calculate_npv(cash_flows, r)
            if abs(npv) < 1:
                return r
        return None


# Step 4: Run calculations
npv_result = calculate_npv(cash_flows, rate)
irr_result = calculate_irr(cash_flows)


# Step 5: Output results
print("\nResults:")
print("NPV:", round(npv_result, 2))

if irr_result is not None:
    print("IRR:", round(irr_result * 100, 2), "%")
else:
    print("IRR: Could not be calculated")


# Step 6: Decision
if npv_result > 0:
    print("GOOD INVESTMENT — NPV is POSITIVE. Your return exceeds the discount rate. Accept this project!")
else:
    print("POOR INVESTMENT — NPV is NEGATIVE. Returns fall short of required rate. Reject or renegotiate")
```
